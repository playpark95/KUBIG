## Crawling #1. Using BeautifulSoup

```python
from bs4 import BeautifulSoup
import urllib.request as req
from urllib.parse import quote

import time
from time import sleep
from random import randint 
```

```python
pages=range(1,7)
```

```python
reviews=[]
authors=[]
stars=[]

for page in pages:
    url = "https://www.amazon.com/COSRX-One-Step-Moisture-Pads/product-reviews/B01N5SHH8Y/ref=cm_cr_getr_d_paging_brm_next_2?ie=UTF8&&reviewerType=all_reviews&PageNumber="+str(page)
    time.sleep(randint(3,6))
    html=req.urlopen(url)
    doc = BeautifulSoup(html, "html.parser")
    
    end1 = len(doc.find_all(class_="a-size-base review-text review-text-contest"))
    reviews.extend([doc.find_all(class_="a-size-base review-text review-text-content")[n].text for n in range(0,end1)])
```

```python
tfidf_uni = TfidfVectorizer(lowercase=True, stop_words=nltk_stopwords, ngram_range=(1,1)).build_analyzer()
```

```python
def tfidf_lem(x):
    return(lem.lemmatize(w) for w in tfidf_uni(x))
```

```python
tfidf_unigram = TfidfVectorizer(analyzer=tfidf_lem)
```

```python
tfidf_review = tfidf_unigram.fit_transform(menu_stripped)
```
